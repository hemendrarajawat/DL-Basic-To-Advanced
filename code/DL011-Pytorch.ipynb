## Numpy <> Pytorch

### Numpy Array to Torch Tensor

```python
a_np = np.random.randn(2, 2)
print(a_np, type(a_np))

a_tensor = torch.from_numpy(a_np)
print(a_tensor, type(a_tensor))
```

## Automatic Differentiation(AutoGrad)

```python
x = torch.ones(3, 2, requires_grad=True)
print(x)
```

```python
y = x+5
print(y)
```

```python
z = y**2 + 1
print(z)
```

```python
t = torch.sum(z)
print(t)
```

```python
t.backward()
```

```python
print(x.grad)
```

```python
x = torch.ones(3, 2, requires_grad=True)
y = x + 5
y_hat = 1. / (1. + torch.exp(-y))
print(y_hat)
s = torch.sum(y_hat)
print(s)
s.backward()
print(x.grad)
```

```python
x = torch.ones(3, 2, requires_grad=True)
y = x + 5
y_hat = 1. / (1. + torch.exp(-y))
print(y_hat)
a = torch.ones(3, 2) * 5
y_hat.backward(a)
print(x.grad)
```

## Simple Loss Function with AutoGrad

```python
X = torch.linspace(-10, 10, 10)
Y = 3 * X - 2

w = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

learning_rate = 0.02
epochs = 200
losses = []
w_values = []
b_values = []

def model(x):
    return w * x + b

def criterion(Y, Y_pred):
    return ((Y - Y_pred)**2).mean()

with torch.no_grad():
    print(f'Predition of x=5 before training: {model(5).item():.4f} and actual is: {3*5-2}')

for epoch in range(1, epochs+1):
    # Forward Pass
    Y_pred = model(X)
    loss = criterion(Y, Y_pred)

    # Backward Pass
    loss.backward()

    # Updating Parameters
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

        w.grad.zero_()
        b.grad.zero_()

        losses.append(criterion(Y, model(X)))
        w_values.append(w.item())
        b_values.append(b.item())

with torch.no_grad():
    print(f'Predition of x=5 after training: {model(5).item():.4f} and actual is: {3*5-2}')

plt.plot(losses, 'r')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.show()
plt.plot(w_values, label='w')
plt.plot(b_values, label='b')
plt.xlabel('Epochs')
plt.ylabel('Value')
plt.legend()
plt.show()
```
