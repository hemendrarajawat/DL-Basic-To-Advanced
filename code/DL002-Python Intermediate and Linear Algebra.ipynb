# Vectors

Vector is a collection of coordinates a point has in a given space.

Vector has both magnitude and direction.

In geometery, a two or more dimension space is called a Euclidean space. A space in any finite no. of dimensions, in which points are designated by coordinates(one for each dimension) and the distance b/w two points is given by a distance formula. L2 norm is also called Euclidean norm(Magnitude).

Magnitude of a vector is given by: $\large \sqrt{\sum_{i=1}^{N} x_i^2}$

## Plotting Vectors

```python
import numpy as np
import matplotlib.pyplot as plt

plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
# Plot multiple vectors
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-4,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
# Creating a method to plot multiple vectors
def plot_vectors(vectors):
    colors = [
        'r', 'y', 'b', 'g', 'c', 'm', 'tan', 'black', 'darkorange', 'limegreen',
        'aqua', 'violet', 'pink', 'magenta', 'teal', 'indigo'
    ]
    
    i = 0
    for vector in vectors:
        plt.quiver(0, 0, vector[0], vector[1], scale_units='xy', angles='xy', scale=1, color=colors[i%len(colors)], label=vector)
        i += 1

    plt.xlim(-10, 10)
    plt.ylim(-15, 15)
    plt.legend()
    plt.show()
```

```python
vectors = np.array([[4,3], [-4,3], [7,1], [3,6]])
plot_vectors(vectors)
```

## Vectors Addition and Substraction

```python
# Addition of two vectors
print(vectors[0] - vectors[1])
plot_vectors([vectors[0], vectors[1], vectors[0] + vectors[1]])
```

```python
# Subtraction of two vectors
print(vectors[0] - vectors[2])
plot_vectors([vectors[0], vectors[2], vectors[0] - vectors[2]])
```

## Vector Dot Product

$\large{\vec{a}\cdot\vec{b} = |\vec{a}| |\vec{b}| \cos(\theta) = a_x b_x + a_y b_y} = a^T b$

```python
print(vectors[0], vectors[2])
dot_product = np.dot(vectors[0], vectors[2])

print(dot_product)
```

## Projection of one vector(a) on another vector(b)

$\large{a_b = |\vec{a}| \cos{\theta} = |\vec{a}| \frac{\vec{a} \cdot \vec{b}}{|\vec{a}| |\vec{b}|} = \frac{\vec{a} \cdot \vec{b}}{|\vec{b}|}}$

$\large{ \vec{a_b} = a_b \hat{b} = a_b \frac{\vec{b}}{|\vec{b}|} }$

```python
a = vectors[0]
b = vectors[2]

plot_vectors([a, b])

a_b = np.dot(a, b)/np.linalg.norm(b)
print('Magnitude of projected vector:', a_b)

vec_a_b = (a_b/np.linalg.norm(b))*b
print('Projected vector:', vec_a_b)

plot_vectors([a, b, vec_a_b])
```

```python
# Another example
a = vectors[1]
b = vectors[2]

plot_vectors([a, b])

a_b = np.dot(a, b)/np.linalg.norm(b)
print('Magnitude of projected vector:', a_b)

vec_a_b = (a_b/np.linalg.norm(b))*b
print('Projected vector:', vec_a_b)

plot_vectors([a, b, vec_a_b])
```

A matrix is a collection of vectors.

## Multiplying a matrix with a vector

The vector gets transformed into a new vector(it strays from its path).

```python
matrix = np.asarray([[1, 2], [2, 4]])
vector = np.asarray([3, 1]).reshape(-1, 1)

# plot_vectors([vector])

print(matrix)
print(vector)

new_vector = np.dot(matrix, vector)
print(new_vector)

plot_vectors([vector, new_vector])
```

## Matrix Addition and Substraction

```python
matrix_1 = np.asarray([
    [1, 0, 3],
    [3, 1, 1],
    [0, 2, 5]
])
matrix_2 = np.asarray([
    [0, 1, 2],
    [3, 0, 5],
    [1, 2, 1]
])
```
